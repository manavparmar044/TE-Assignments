```python
idf = pd.read_csv("Iris.csv")
idf.head()
```

```python
setosa = idf[idf['Species'] == 'Iris-setosa']
versicolor = idf[idf['Species'] == 'Iris-versicolor']
virginica = idf[idf['Species'] == 'Iris-virginica']
```

```python
print("📊 Statistics for Iris-setosa:")
print(setosa.describe())
print("\n📊 Statistics for Iris-versicolor:")
print(versicolor.describe())
print("\n📊 Statistics for Iris-virginica:")
print(virginica.describe())
```
