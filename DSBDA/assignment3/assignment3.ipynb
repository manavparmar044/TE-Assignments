```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nba.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["College"] = df["College"].fillna(df["College"].mode())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
height_groups = df.groupby(df["Height"])
height_groups.head()
```

```python
height_groups["Salary"].describe()
```

```python
df["Age"].max()
```

```python
bins = [19,25,31,36,40]
labels = labels = ["19-24", "25-30", "31-35", "36-40"]
df["AgeGroup"] = pd.cut(df["Age"], bins = bins, labels = labels)
```

```python
df["AgeGroup"].value_counts()
```

```python
age_groups = df.groupby("AgeGroup")
```

```python
age_groups["Salary"].describe()
```
