```python
sns.histplot(data = df, x = "SepalLengthCm", bins = 15, kde = True)
```

```python
sns.histplot(data = df, x = "SepalWidthCm", bins = 15, kde = True)
```

```python
sns.histplot(data = df, x = "PetalLengthCm", bins = 15, kde = True)
```
