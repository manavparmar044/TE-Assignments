```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.isnull().sum()
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
df["embark_town"] = df["age"].ffill()
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.boxplot(data = df, x = "sex", y = "age", hue = "survived")
```

```python
encoder = LabelEncoder()
df["sex_enc"] = encoder.fit_transform(df["sex"])
df["embarked_enc"] = encoder.fit_transform(df["embarked"])
```

```python
sns.heatmap(df.drop(["alive","alone","who","adult_male","sex","class","deck","embarked"], axis=1).corr(), annot=True)
```
